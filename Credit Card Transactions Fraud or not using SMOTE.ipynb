```python
import tensorflow as tf
import numpy
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.listdir()
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.Class.value_counts()
```

```python
df.isnull().sum()
```

```python
df2 = df.drop(['Time','Amount'],axis = 1)
df2.head()
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
standard_scaler = StandardScaler()
scaled_amount = standard_scaler.fit_transform(np.array(df.Amount).reshape(-1,1))
df2['norm_amount'] = scaled_amount
df2.head()
```

```python
df3 = df2.drop('Class',axis = 1)
df3.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(df3,df2['Class'],stratify=df2['Class'],shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report
y_pred = lr.predict(X_test)
print(classification_report(Y_test,y_pred))
```

```python
df_not_fraud = df2[df2['Class'] ==  0]
df_fraud = df2[df2.Class == 1]
df_fraud.head()
```

```python
df_not_fraud.head()
```

```python
df_not_fraud_undersampled = df_not_fraud.sample(df_fraud.shape[0])
df_not_fraud_undersampled.shape
```

```python
df4 = pd.concat([df_fraud,df_not_fraud_undersampled],axis = 0)
df4.head()
```

```python
df4.drop('Class',axis = 1).head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(df4.drop('Class',axis = 1),df4['Class'],stratify=df4['Class'],shuffle=True)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,Y_train)
y_pred = lr.predict(X_test)
print(classification_report(Y_test,y_pred))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
sns.heatmap(cm,annot=True,fmt=".1f")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,oob_score = True,criterion='entropy')
rf.fit(X_train,Y_train)
y_pred = rf.predict(X_test)
print(classification_report(Y_test,y_pred))
```

```python
features_importance = rf.feature_importances_
forest_importances = pd.Series(features_importance, index=df4.drop('Class',axis = 1).columns)

fig, ax = plt.subplots()
forest_importances.plot.bar(forest_importances)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(df4.drop('Class',axis = 1),df4['Class'],stratify=df4['Class'],shuffle=True)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)
print(classification_report(Y_test,y_pred))
```

```python
model.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
model = GradientBoostingClassifier()
params = {'max_depth':[3,4,5,6],
         'max_features':[0.2,0.4,0.6,0.8]}
GSCV = GridSearchCV(param_grid=params,cv=10,estimator=model)
GSCV.fit(X_train,Y_train)
```

```python
GSCV.best_params_
```

```python
estimator = GSCV.best_estimator_
y_pred = estimator.predict(X_test)
print(classification_report(Y_test,y_pred))
```
